```python
pd.set_option('display.max.colwidth', 100)

df_nts = pd.DataFrame(js_notice)
df_nts
```

### 업데이트 내역 전처리

```python
df = df_nts[df_nts['Title'].str.contains('업데이트', na=False)]
df = df.sort_values('Date')
df['Date'] = df['Date'].str[:10]
df = df.drop(['Type'], axis=1)
df.reset_index(drop=True, inplace=True)
```

```python
df, df.info()
```

### 누락된 6월 21일 업데이트 내역 추가

```python
df_0621 = df_nts[df_nts['Date'].str.contains('2023-06-21', na=False)]
df_0621_2 = df_0621.copy()

df_0621_2['Date'] = df_0621_2['Date'].str[:10]
df_0621_2 = df_0621_2[df_0621_2['Title'].str.contains('점검', na=False)]
df_0621_2 = df_0621_2.drop(['Type'], axis=1)
df_0621_2.reset_index(drop=True, inplace=True)
df_0621_2
```

```python
df_0517_0802 = pd.concat([df, df_0621_2], ignore_index=True)
df_0517_0802 = df_0517_0802.sort_values('Date')

df_0517_0802['Date'] = pd.to_datetime(df_0517_0802['Date'])
df_0517_0802
```

```python
df_0517_0802.to_excel('df_0517_0802.xlsx', index=False)
notice_df = pd.read_excel('df_0517_0802.xlsx')
notice_df
```
